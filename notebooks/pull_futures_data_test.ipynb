```python
summary_table = pd.DataFrame(columns=[
    "Commodity",
    "Contract Code",
    "N", 
    "Basis", 
    "Freq. of Backwardation (%)", 
    "E(Re) (Mean Annual Excess Return)", 
    "σ(Re) (Std Dev of Excess Return)", 
    "Sharpe Ratio"
])
```

```python
for product_contract_code in [3160, 289, 3161, 1980, 2038, 3247, 1992, 361, 385, 2036, 379, 3256, 396, 430,
                              1986, 2091, 2029, 2060, 3847, 2032,
                              3250, 2676, 2675,
                              3126, 2087, 2026, 2020, 2065, 2074, 2108]:
    # Step 1: Get relevant futcodes for product
    query_info = f"""
    SELECT futcode, contrcode, contrname, startdate, lasttrddate
    FROM tr_ds_fut.wrds_contract_info
    WHERE contrcode = {product_contract_code}
    AND startdate >= '1970-01-01' AND lasttrddate <= '2008-12-31'
    """
    data_info = db.raw_sql(query_info)
    if len(data_info) == 0:
        print("No data available for the specified product. Product Code:", product_contract_code)
        continue

    
    # Extract relevant futcodes
    ng_futcodes = tuple(data_info["futcode"].unique())

    # Step 2: Get settlement prices for NG contracts
    query_contracts = f"""
    SELECT futcode, date_, settlement
    FROM tr_ds_fut.wrds_fut_contract
    WHERE futcode IN {ng_futcodes}
    AND date_ >= '1970-01-01' AND date_ <= '2008-12-31'
    """
    data_contracts = db.raw_sql(query_contracts)

    # Convert date column to datetime
    data_contracts["date_"] = pd.to_datetime(data_contracts["date_"])

    # Extract year-month for grouping
    data_contracts["month"] = data_contracts["date_"].dt.to_period("M")

    # Step 3: Get prompt & 12-month futures settlement prices
    def get_prompt_and_12m(group):
        """Extract nearest and 12-month maturity contract settlement prices for each month."""
        group = group.sort_values(by="date_")
        prompt_contract = group.iloc[0]  # First contract (nearest)
        contract_12m = group.iloc[-1] if len(group) > 1 else None  # Last contract (12 months out)
        
        return pd.Series({
            "prompt_settlement": prompt_contract["settlement"],
            "contract_12m_settlement": contract_12m["settlement"] if contract_12m is not None else None
        })

    # Apply function to get prompt & 12-month contracts per month
    monthly_contracts = data_contracts.groupby("month").apply(get_prompt_and_12m, include_groups=False).reset_index()

    # Step 4: Compute the basis
    monthly_contracts["basis"] = np.log(monthly_contracts["prompt_settlement"]) - np.log(monthly_contracts["contract_12m_settlement"])
    monthly_contracts["basis"] /= 11  # Annualized basis

    # Step 5: Compute frequency of backwardation
    freq_backwardation = (monthly_contracts["basis"] > 0).mean() * 100

    # Compute excess return for the prompt contract
    monthly_contracts["excess_return"] = monthly_contracts["prompt_settlement"].pct_change() * 100

    # Compute mean and std deviation of annualized excess return
    excess_return_mean = monthly_contracts["excess_return"].mean()
    excess_return_std = monthly_contracts["excess_return"].std()

    # Compute Sharpe Ratio
    sharpe_ratio = excess_return_mean / excess_return_std if excess_return_std != 0 else np.nan

    # Count number of valid observations
    N = len(monthly_contracts.dropna())

    summary_table = pd.concat([summary_table, pd.DataFrame({
        "Commodity": [data_info['contrname'].unique()[0]],
        "Contract Code": [data_info['contrcode'].unique()[0]],
        "N": [N],
        "Basis": [monthly_contracts["basis"].mean()],
        "Freq. of Backwardation (%)": [freq_backwardation],
        "E(Re) (Mean Annual Excess Return)": [excess_return_mean],
        "σ(Re) (Std Dev of Excess Return)": [excess_return_std],
        "Sharpe Ratio": [sharpe_ratio]
    })], ignore_index=True)


```
